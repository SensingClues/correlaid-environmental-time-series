```r
source("../scripts/utils.R")
source("../scripts/data_processing.R")
load_libraries()
```

```r
# Set paths and load data for Zambia
country_name <- "Zambia"
resolution <- 1000  # example resolution
data_type <- "NDVI"  # example data type
crs <- "EPSG:4326"

paths_data <- set_paths_and_load_data(country_name, resolution, data_type)

# Extract and order dates for historic data
historic_dates_info <- extract_and_order_dates(paths_data$ndvi_files_historic)

# Extract and order dates for current data
current_dates_info <- extract_and_order_dates(paths_data$ndvi_files_current)

# Load, transform, and mask historic data
historic_data <- load_transform_and_mask_data(paths_data$data_path_historic, historic_dates_info$ndvi_files, historic_dates_info$dates, paths_data$aoi_path, paths_data$aoi_files, crs)

# Load, transform, and mask current data
current_data <- load_transform_and_mask_data(paths_data$data_path_current, current_dates_info$ndvi_files, current_dates_info$dates, paths_data$aoi_path, paths_data$aoi_files, crs)

# Inspect NDVI values distribution for historic data
historic_plots <- inspect_ndvi_distribution(historic_data$ndvi_msk, historic_dates_info$dates, country_name)

# Inspect NDVI values distribution for current data
current_plots <- inspect_ndvi_distribution(current_data$ndvi_msk, current_dates_info$dates, country_name)

# Combine historic and current NDVI data
combined_data <- combine_ndvi_data(historic_data$ndvi_msk, current_data$ndvi_msk, historic_dates_info$dates, current_dates_info$dates)

# Plot combined NDVI data
combined_ndvi_plots <- plot_combined_ndvi(combined_data, country_name)
```

```r
# Compute and plot NDVI maps for historic data
compute_and_plot_ndvi_maps(historic_data$ndvi_msk, historic_dates_info$dates)
```
